#### サブプロット
- fig, ax
- subplots, add_subplot

```python
# データの作成
Y1 = [61, 45, 27, 88, 47]
Y2 = [17, 39, 46, 40, 28]
labels = ["A", "B", "C", "D","E"]

# figureとaxesを作る
# サブプロットとは、グラフの枠を複数確保し、管理するためのメソッドである
# subplot"s"はaxesをまとめてつくるためのメソッド plt.subplotはひとつづつaxesを作る。
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2) # 1行2列の型

# 1行2列の左
ax1.bar(labels, Y1, color="b")
ax1.set_title("dog")

# 1行2列の右
ax2.bar(labels, Y2, color="g")
ax2.set_title("cat")

plt.show()
```

```python
# データの作成
Y1 = [61, 45, 27, 88, 47]
Y2 = [17, 39, 46, 40, 28]
labels = ["A", "B", "C", "D","E"]

# figureを作る
fig = plt.figure()

# 1行2列の左
# add_subplotは、空のfigureにサブプロットを追加するためのメソッド
# plt.subplotでも同じように作れるが、直近のFigureを自動的に選ぶため、どのFigureを編集したいかコントロールができない
ax1 = fig.add_subplot(1, 2, 1) # axesを作る
ax1.bar(labels, Y1, color="b")
# サブプロットの枠ごとにタイトルを描画したいときは、axes.set_titleを用いる
ax1.set_title("dog")

# 1行2列の右
ax2 = fig.add_subplot(1, 2, 2) # axesを作る
ax2.bar(labels, Y2, color="g")
ax2.set_title("cat")

plt.show()
```

#### グラフを上下に並べる
- tight_layout

```python
# データの作成
Y1 = [61, 45, 27, 88, 47, 56, 61]
Y2 = [17, 39, 46, 40, 27, 35, 41]
labels = ["A", "B", "C", "D","E", "F", "G"]

# figureを作る
fig = plt.figure()

# 2行1列の上
ax1 = fig.add_subplot(2, 1, 1)
ax1.bar(labels, Y1, color="b")
ax1.set_title("snake")

# 2行1列の下
ax1 = fig.add_subplot(2, 1, 2)
ax1.bar(labels, Y2, color="g")
ax1.set_title("fish")

# 図の重複を避ける
plt.tight_layout()
```

#### y軸のレンジを統一する
- ylim

```python
#　データの作成
Y1 = [61, 45, 27, 88, 47]
Y2 = [17, 39, 46, 40, 27]
labels = ["A", "B", "C", "D", "E"]

# figureを作る
fig = plt.figure()

# 1行2列の左
ax1 = fig.add_subplot(1, 2, 1)
ax1.bar(labels, Y1, color="b")
ax1.set_title("dog")
ymin, ymax = ax1.get_ylim()  # ax1のy軸のレンジを取得する

# 1行2列の右
ax2 = fig.add_subplot(1, 2, 2)
ax2.bar(labels, Y2, color="g")
ax2.set_title("cat")
ax2.set_ylim(ymin, ymax) # y軸のレンジをax1と合わせる

plt.show()
```

#### ２軸グラフ
- twinx

```python
#　データの作成
X1, Y1 = range(0, 5), [61, 45, 27, 88, 47]
X2, Y2 = [x + 0.3 for x in range(0, 5)], [17, 39, 46, 40, 27]
X3, Y3 = [x + 0.6 for x in range(0, 5)], [17, 39, 46, 40, 27]
labels = ["A", "B", "C", "D", "E"]
print("X2: " + str(X2))

# figureを作る
fig = plt.figure()

# 1行1列の左軸
ax1 = fig.add_subplot(1, 1, 1)
ax1.bar(X1, Y1, color="b", tick_label=labels, width=0.3) 

# 1行1列の右軸
ax2 = ax1.twinx()
ax2.bar(X2, Y2, color="g", tick_label=labels, width=0.3)

ax3 = ax1.twinx()
ax3.bar(X3, Y3, color="r", tick_label=labels, width=0.3)
plt.show()
```

```python
#　データの作成
X1, Y1 = range(0, 5), [61, 45, 27, 88, 47]
X2, Y2 = range(0, 5), [17, 39, 46, 40, 27]
labels = ["A", "B", "C", "D", "E"]

# figureを作る
fig = plt.figure()

# 1行1列の左軸
ax1 = fig.add_subplot(1, 1, 1)
ax1.bar(X1, Y1, color="b", tick_label = labels, width=0.4, label="left") 

# 1行1列の右軸
ax2 = ax1.twinx()
ax2.plot(X2, Y2, color="g", label="right")

# 凡例を設定
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
plt.legend(handler1+handler2, label1+label2)

plt.show()
```

#### figsize and complex layout

```python
# データの作成
Y1 = [61, 45, 27, 88, 47, 56, 61]
Y2 = [17, 39, 46, 40, 27, 35, 41]
Y3 = [1, 2, 3, 4, 5, 6, 7]
Y4 = [8, 9, 10, 11, 12, 13, 14]
Ys = [Y1, Y2, Y3, Y4]
titles = ["snake", "fish", "Y3", "Y4"]
labels = ["A", "B", "C", "D","E", "F", "G"]

# figureを作る
fig = plt.figure(figsize=(10,10))

for i,y in enumerate(Ys):
    ax = fig.add_subplot(2,2,i+1)
    ax.bar(labels, y)
    ax.set_title(titles[i])
    ax.set_ylim(0,61)

# 図の重複を避ける
# plt.tight_layout()
plt.show()
```

#### multiple figures

```python
import numpy as np
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2

# figure 1 つ目
plt.figure(figsize=(3, 3))
plt.subplot(1,3,3)
plt.plot(x, y1)
plt.title("3rd")
plt.subplot(1,3,2)
plt.plot(x, y1)
plt.ylim(-3,8)
plt.title("2nd")
plt.subplot(1,3,1)
plt.plot(x, y2)

# figure 2 つ目
plt.figure(figsize=(4, 4))
plt.plot(x, y2)
plt.show()
```
